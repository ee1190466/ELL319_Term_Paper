```python
import pandas as pd
import numpy as np
import os
import math
from scipy import signal
import scipy.io
import mne
```

```python
lenarr = []
eeg = pd.read_csv('dataICA.txt',sep='\t')
eeg = eeg.to_numpy()
eeg = eeg[:,1:32]
event_file = pd.read_csv('events.txt',sep='\t')


timestamp = event_file['latency'].to_numpy()
event = event_file['type'].to_numpy()

event_arr = []
timestamp_arr = []

for i in range(0,len(event)):
    if event[i]!='boundary':
        event_arr.append(event[i])
        timestamp_arr.append(timestamp[i])

for i in range(0,len(event_arr)):
    
    if event_arr[i]!='PP':
    
        start = timestamp_arr[i]
        stop = timestamp_arr[i+1]
        eeg_sig = eeg[int(start-500):int(stop),:]
        trial = int(i//52)
        print(trial)
        np.save('Segmented/ICA_'+event_arr[i]+'_'+str(trial)+'.npy',eeg_sig)
        print(len(eeg_sig))
        lenarr.append(len(eeg_sig))
```
